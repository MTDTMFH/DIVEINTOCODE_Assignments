# ダミーデータの作成

機械学習に関するプログラムの実験を行う上で、実データではないランダムなデータ（ダミーデータ）が欲しいということがあります。正規分布に従う乱数によるダミーデータの作成を通じてNumPyの使い方になれていきましょう。


散布図で描くと次のようになるデータを作っていきます。

### 【問題1】乱数の作成

平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

```python
import numpy as np
mean = np.array([-3, 0])
covariance = np.array([[1, 0.8], [0.8, 1]])

print(mean.shape, covariance.shape)
```

```python
np.random.seed(0)
x = np.random.multivariate_normal(mean, covariance, 500)
x
```

```python
# set x1
x1 = x[:, 0]
x1
```

```python
# set x2
x2 = x[:, 1]
x2
```

```python
# mean check x1
x1_mean = np.mean(x[:, 0])
x1_mean
```

```python
# mean check x2
x2_mean = np.mean(x[:, 1])
x2_mean
```

```python
# covariance check

cov = np.sum(((x1 - x1_mean)*(x2 - x2_mean))) / 500
cov
```

```python
cov_x1 = np.sum(((x1 - x1_mean)*(x1 - x1_mean))) / 500
cov_x1
```

```python
cov_x2 = np.sum(((x2 - x2_mean)*(x2 - x2_mean))) / 500
cov_x2
```

### 【問題2】散布図による可視化

問題1で作成したデータ点を散布図により可視化してください。


散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
import matplotlib.pyplot as plt

plt.title("Scatter Plot of Pseudo Random Variables")
plt.ylabel("x2")
plt.xlabel("x1")
plt.scatter(x1, x2)
plt.show()
```

### 【問題3】ヒストグラムによる可視化

問題1で作成したデータをヒストグラムにより可視化してください。


ヒストグラムはplt.hist()を使うことで描けます。

各次元ごとに作成し、2つのヒストグラムとしてください。

それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。

```python
# x1
plt.title("Histogram of x1")
plt.ylabel("frequency")
plt.xlabel("x1")
plt.hist(x1, bins=50)
plt.xlim([-5, 2])
plt.show()
```

```python
# x2
plt.title("Histogram of x2")
plt.ylabel("frequency")
plt.xlabel("x2")
plt.hist(x2, bins=50)
plt.xlim([-5, 2])
plt.show()
```

### 【問題4】データの追加

新たに平均が(0, -3)、共分散行列が上記（問題１）同様の2次元正規分布による乱数を500個作成してください。

そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題4のものは1を表示してください。

```python
mean_2 = [0, -3]

np.random.seed(1)
new = np.random.multivariate_normal(mean_2, covariance, 500)

new_x1 = new[:, 0]
new_x2 = new[:, 1]
```

```python
plt.title("Scatter Plot of Pseudo Random Variables")
plt.ylabel("x2")
plt.xlabel("x1")
zero = plt.scatter(x1, x2)
one = plt.scatter(new_x1, new_x2)
plt.legend((zero, one), ("0", "1"), loc='upper right')
plt.show()
```

### 【問題6】ラベル付け

ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。


0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。


機械学習に使用するデータセットはこのような形になっていることが多いです。


《作成例》


データが3個+3個の場合は以下のような(6, 3)のndarrayになります。

```python
# Create labels
first_500 = np.zeros(500)
second_500 = np.ones(500)
#print(first_500.shape, second_500.shape)
#print(first_500, second_500)

labels = np.concatenate([first_500, second_500])
```
