### 【問題1】べき乗の算術演算子を使用して作成

今回の課題では 1枚の紙を43回折り曲げた時の厚さを計算するプログラム を作成します。
折る前の紙の厚さ t0 は0.00008mとします。一般的なコピー用紙の厚さです。
べき乗の算術演算子を使用したプログラムを作ってください。

```python
THICKNESS = 0.00008
Times_to_fold = 43

def thickness_after_fold(Thickness, no_fold):
    folded_thickness = Thickness * (2 ** no_fold)
    return folded_thickness

folded_thickness = thickness_after_fold(THICKNESS, Times_to_fold)
print("{}回折った厚さ： {}メートル".format(Times_to_fold, folded_thickness))
```

### 【問題3】for文を使用して作成

次に、for文を使用したプログラムを作ってください。
べき乗の算術演算子は使ってはいけません。算術演算子は四則演算（+、-、*、/）のみ使えます。

```python
exponent_base = 1
Times_to_fold = 43

for i in range(Times_to_fold):
    exponent_base *= 2
    folded_thickness = (exponent_base * THICKNESS)
print("{}回折った厚さ： {}メートル".format(Times_to_fold, folded_thickness))
```

### 【問題5】リストへの保存

ここまでは43回折った後の最後の値だけを使用していましたが、グラフで可視化するためには過程の値も必要です。for文を使用したコードに、過程の値合計44個を記録するコードを加えてください。


《ヒント》

- 空のリストを作成する。
- 折る前の値をリストに追加する。
- for文の中でn回折った時の値をリストに追加していく。

<br> 最終的にリストに44個の値が格納されていることをlen関数を用いて確認しておきましょう

```python
iterations = []
values = []
combined = []
Times_to_fold = 43
THICKNESS = 0.0008
exponent_base = 1

for i in range(Times_to_fold):
    exponent_base *= 2
    folded_thickness = (exponent_base * THICKNESS)
    print(i+1, folded_thickness)
    iterations.append(i+1)
    values.append(folded_thickness)
print("{}回折った厚さ： {}メートル".format(Times_to_fold, folded_thickness))

combined = zip(iterations, values)
combined_results = list(combined)
print(combined_results)
```

### 【問題6】折れ線グラフの表示

グラフの描画には Matplotlib という ライブラリ を用います。リストへ記録するコードの後ろで以下の雛形を使用してください。

```python
"""
グラフを表示する。タイトルと軸ラベル名付き。
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Thickness of folded paper")
plt.xlabel("Number of folds")
plt.ylabel("Thickness[m]")
plt.plot(iterations, values) 
plt.show()
```
