###  問題1】train_test_splitのスクラッチ

スクラッチの練習として、scikit-learnのtrain_test_splitを自作してみます。以下の雛形をベースとして関数を完成させてください。

[sklearn.model_selection.train_test_split — scikit-learn 0.21.3 documentation](https://diver.diveintocode.jp/curriculums/1643)

なお、作成した関数がscikit-learnのtrain_test_splitと同じ動作をしているか必ず確認をするようにしましょう。


```python
import numpy as np
import pandas as pd
import random
import math
```

```python
def scratch_train_test_split(X, y, train_size=0.8,):
    """
    検証データを分割する。
    Parameters
    ----------
    X : 次の形のndarray, shape (n_samples, n_features)
      訓練データ
    y : 次の形のndarray, shape (n_samples, )
      正解値
    train_size : float (0<train_size<1)
      何割をtrainとするか指定
    Returns
    ----------
    X_train : 次の形のndarray, shape (n_samples, n_features)
      訓練データ
    X_test : 次の形のndarray, shape (n_samples, n_features)
      検証データ
    y_train : 次の形のndarray, shape (n_samples, )
      訓練データの正解値
    y_test : 次の形のndarray, shape (n_samples, )
      検証データの正解値
    """
    #ここにコードを書く
    
    #error handling
    if (X.shape[0] != y.shape[0]):
        print("Check the size of inputs - X is {} & y is {}".format(X.shape[0], y.shape[0]))
    
    #if (y.shape >= 1):
        #print("Check the output of y! Should be an 1d array")
        
    #input data
    n_rows = X.shape[0]
    split = train_size * X.shape[0]
    n_size = X.shape[0] * X.shape[1]
    split_X = train_size
    total_X = math.floor(n_rows * split_X)

    #randomiser - X
    X_mix = []
    X_mix = random.sample(list(X.flatten()), n_size)
    
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            X[i][j] = X_mix[0]
            X_mix.pop(0)

    #randomiser - Y
    y_mix = []
    y_mix = random.sample(list(y.flatten()), len(y))
    
    for i in range(len(y)):
            y[i] = y_mix[0]
            y_mix.pop(0)
    
    combined = np.hstack([X, y.reshape(len(y), 1)])
    
    X_train = combined[:total_X, :-1]
    X_test = combined[total_X:, :-1]
    y_train = combined[:total_X, -1]
    y_test = combined[total_X:, -1]
    
    return X_train, X_test, y_train, y_test
```
