### 【問題】栗まんじゅうが太陽系を覆う日

1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。

栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。

ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも適用できるような汎用的なプログラムを作成してください。また、グラフを表示するなど分かりやすいノートブックを作成してください。

````python
import math

# size of the solar system
'''
https://www.universetoday.com/15585/diameter-of-the-solar-system/#:~:text=1%20AU(distance%20from%20the,is%201%2C921.56%20AU%20in%20diameter.
'''
diameter_in_AU = 1921.56

# unit of measurement
'''
天文単位（てんもんたんい、英: astronomical unit、記号: au）は長さの単位で、149597870700 m である。
'''

# Assumption
'''
１）太陽系は球体であると問題を単純化する
太陽系の体積を求める式は以下の通りとする
'''
solar_system_vol = 4/3*math.pi*(1921.56/2)


'''
２）栗饅頭は直方体であると問題を単純化する
栗饅頭の体積を求める式は以下の通りとする
5cm * 3cm * 1cm = l * w * h
0.05m * 0.03m * 0.01m
'''

kurimanjyu_vol_AU = (0.05/149597870700)*(0.03*149597870700)*(0.01/149597870700)

print(solar_system_vol)
print(kurimanjyu_vol)
````

````python
# オブジェクト指向での演習

class Item:
    def __init__(self, name, volume):
        self.name = name
        self.volume = volume
        
class Space:
    def __init__(self, name, volume):
        self.name = name
        self.volume = volume
    

solar_system = Space('太陽系', solar_system_vol)
kurimanjyu = Item('栗饅頭', kurimanjyu_vol_AU)
````

````python
#  栗饅頭 v 太陽系
'''
アプローチとしては、５分間の倍増を何回で特定のスペースを覆うのかを算出して、それを１２回で割ることでX時間を求める
'''

def cover_space(item, space, rate):
    i = 0
    
    while True:
        i+=1
        item_vol = item.volume * (2**i)
        
        if (item_vol >= space.volume):
            break
    print("バイバイんをふりかけることで{}は{}分ごとに倍増し、約{}分で{}を覆う計算になる".format(item.name, rate, i*rate, space.name))
        
cover_space(kurimanjyu, solar_system, 5)
````
